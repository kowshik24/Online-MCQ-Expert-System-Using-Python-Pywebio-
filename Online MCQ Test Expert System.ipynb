```python
from pywebio.input import *
from pywebio.output import *
import random
```

```python
import pandas as pd

df = pd.read_csv(r'F:\6th Semister\Artificial Intellegence\Project_Web\dataset.csv')
```

```python

entered = []
n = 10

questions = random.sample(range(df.shape[0]),n)

def exam():
    name = input("Enter your name : " , type="text")
    roll = input("Enter your roll number: ",type="text")
    
    row = df.shape[0]
    

    c = 0
    for j,i in enumerate(questions):
        options = []
        question = df.iloc[i][0]
        options.append(df.iloc[i][1])
        options.append(df.iloc[i][2])
        options.append(df.iloc[i][3])
        options.append(df.iloc[i][4])
        ans = df['answer'][i]
        q = radio(question,options)
        choose = df[ans][i]
        entered.append(q)
        if q==choose:
            c+=1
            
    
    
    percentage = (c/n)*100
    percent = round(percentage,2)    
    if percentage>=33.33:
        put_text(name+" , your score is "+str(c)+" out of "+ str(n))
        
        put_text("Success Rate : "+ str(percent) + "%")
        #print("Seccess Rate : {:.2f}".format(round(percentage, 2)))
        style(put_text("Result : Passed!!"),"color:green")
    else:
         put_text(name+" , your score is "+str(c)+" out of "+ str(n))
         put_text("Success Rate : "+ str(percent) + "%")
         style(put_text("Result : Failed!!"),"color:red")

    put_text("Thank you!!!")
    


def show_answes_script():
    
    put_text(".............................. || Your Answer Script || ..........................")
    
    row = df.shape[0]
    
    for j,i in enumerate(questions):
        #options = []
        question = df.iloc[i][0]
        q = "Q"+str(i+1)+" "
        put_text(q,question)
        put_text("a) ",df.iloc[i][1])
        put_text("b) ",df.iloc[i][2])
        put_text("c) ",df.iloc[i][3])
        put_text("d) ",df.iloc[i][4])
       
        ans = df['answer'][i]
        #q = radio(question,options)
        choose = df[ans][i]
        put_text("The correct answer is : ", choose)
        put_text("You entered : ",entered[j])
        
        if choose == entered[j]:
            style(put_text("Correct!!"),"color:green")
        else:
            style(put_text("Incorrect!!"),"color:red")
            


if __name__== '__main__':    
    exam()
    show_answes_script()
    


    

    

```
